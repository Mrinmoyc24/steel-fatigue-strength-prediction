### Linear Regression Model

```python
X=df1.drop(columns=['Sl. No.','Fatigue'])
y=df1['Fatigue']
```

```python
#train test split set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
#creating and training the model
from sklearn.linear_model import LinearRegression
lm= LinearRegression()
lm.fit(X_train, y_train)
```

```python
#model evaluation
print(lm.intercept_)
```

```python
coeff_df=pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
#prediction from our model
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('prediction')
plt.ylabel('actual')
#its almost a straight line so our prediction is fair
```

```python
sns.displot((y_test-predictions),bins=50, kde=True)
#almost gaussian curve so our pred almost ok
```

```python
#Regretion evaluation
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
from sklearn.metrics import r2_score
r2_lr = r2_score(y_test, predictions)
print("R-squared:", r2_lr)

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions, color='blue', edgecolor='k', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', linewidth=2)

plt.xlabel("Actual Fatigue Strength")
plt.ylabel("Predicted Fatigue Strength")
plt.title("LRM Regression: Actual vs Predicted")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### XGBoost

```python
#now we need to define our xtest and y test again as it was been scaled before
X=df1.drop(columns=['Sl. No.','Fatigue'])
y=df1['Fatigue']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error

# Initialize the model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)

# Train the model
xgb_model.fit(X_train, y_train)

# Predict on test set
xgb_predictions = xgb_model.predict(X_test)

# Evaluation
r2 = r2_score(y_test, xgb_predictions)
mse = mean_squared_error(y_test, xgb_predictions)

print("XGBoost R-squared:", r2)
print("XGBoost Mean Squared Error:", mse)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, xgb_predictions, color='blue', edgecolor='k', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', linewidth=2)

plt.xlabel("Actual Fatigue Strength ")
plt.ylabel("Predicted Fatigue Strength ")
plt.title("XGB Regression: Actual vs Predicted")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mae = metrics.mean_absolute_error(y_test, xgb_predictions)
mse = metrics.mean_squared_error(y_test, xgb_predictions)
rmse = np.sqrt(mse)
r2_XGB = r2_score(y_test, xgb_predictions)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared:", r2_XGB)
```
